```python
review = review_html.find_all("div",{"class": "detailed-reviews-userReviewsContainer"})
```

```python
review
```

```python
for i in review:
    user_rating = i.findAll("div",{"class":"user-review-main user-review-showRating"})
    user_comment= i.findAll("div",{"class":"user-review-reviewTextWrapper"})
    user_name = i.findAll("div",{"class":"user-review-left"})

```

```python
user_rating
```

```python
user_comment
```

```python
user_name
```

```python
# Loop through the length of user ratings
for i in range(len(user_rating)):
    # Extract the rating text and remove extra spaces
    rating = user_rating[i].find("span", {"class": "user-review-starRating"}).get_text().strip()
    
    # Extract the user comment text
    comment = user_comment[i].text.strip()  # Ensure to remove extra spaces

    # Extract the user name (assuming the name is inside a <span>)
    name = user_name[i].find("span").get_text().strip()

    # Extract the date (assuming it's the second <span> element)
    date = user_name[i].find_all("span")[1].get_text().strip()

    # Print the extracted information in a formatted string
    print(f"Date: {date}, Rating: {rating}, Name: {name}, Comment: {comment}")
```
